```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

trans_data = movie_data.copy()
# Transform Director columns
trans_data['Name'] = labelencoder.fit_transform(movie_data['Name'])
trans_data['Director'] = labelencoder.fit_transform(movie_data['Director'])

# Transform Actors Columns
trans_data['Actor 1'] = labelencoder.fit_transform(movie_data['Actor 1'])
trans_data['Actor 2'] = labelencoder.fit_transform(movie_data['Actor 2'])
trans_data['Actor 3'] = labelencoder.fit_transform(movie_data['Actor 3'])
trans_data['Genre'] = labelencoder.fit_transform(movie_data['Genre'])


trans_data.head()
```

```python
# Import modeling Libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
# Function of Multiple Linear Regression
def mlr(X, y):
    # Add a column of ones to X for the intercept term
    ones = np.ones((len(X),1))
    X = np.hstack((ones, X))

    # start to fomulateing with this formula : (X^T * X)^-1 * (X^T * y)

    # transpose of the X
    X_transpose = np.transpose(X)

    # X^T * X
    X_transpose_X = np.dot(X_transpose, X)

    # inverse of X^T * X
    X_transpose_X_inv = np.linalg.pinv(X_transpose_X)

    # then calculate the X^T * y
    X_transpose_y = np.dot(X_transpose, y)

    # Multiply the two term
    beta = np.dot(X_transpose_X_inv, X_transpose_y)

    return beta

# Function to predict using the fitted model
def predict_mlr(X, beta):

    # Step 1: Add a column of ones to X for the intercept term

    # Create a column of ones
    ones = np.ones((len(X), 1))
    X = np.hstack((ones, X))

    # Step 2: Compute predictions

     # Multiply X and beta
    y_pred = np.dot(X, beta)

    return y_pred

# Example dataset (features and target)
X = trans_data.drop(['Rating'],axis=1)
y = trans_data['Rating']


# Fit the MLR model
beta = mlr(X, y)
print("Coefficients (including intercept):", beta)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_test = np.array(X_test)
y_pred = predict_mlr(X_test, beta)
y_pred
```

```python
print('R2 score: ',r2_score(y_test,y_pred))
print('Mean squared error: ',mean_squared_error(y_test,y_pred))
print('Mean absolute error: ',mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
randregressor=RandomForestRegressor(n_estimators=50)
randregressor.fit(X_train,y_train)
predictions = randregressor.predict(X_test)
print('R2 score: ',r2_score(y_test,predictions))
print('Mean squared error: ',mean_squared_error(y_test,predictions))
print('Mean absolute error: ',mean_absolute_error(y_test,predictions))
```
